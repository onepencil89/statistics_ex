# 분포 통계량

```python
import numpy as np
from scipy import stats
import pandas as pd
```

## 백분위 수 및 사분위수 계산

```python
x = np.arange(1, 12, 1)
print(x)
```

```python
print(np.percentile(x, 20))
print(np.quantile(x, 0.2))
```

## 왜도 계산

```python
x1 = [1] * 30 + [2] * 20 + [3] * 20 + [4] * 15 + [5] * 15 # 좌측으로 치우침
x2 = [1] * 15 + [2] * 20 + [3] * 30 + [4] * 20 + [5] * 15 # 치우치지 않음
x3 = [1] * 15 + [2] * 15 + [3] * 20 + [4] * 20 + [5] * 30 # 우측으로 치우침
```

```python
len(x1)
```

```python
pd_s = pd.Series(x1)
```

```python
#vector의 시각화
pd_s.value_counts(ascending=True)

#s ascending은 오름차순, count의 오름차순(True)
```

```python
pd_s.value_counts(ascending=False)

#ascending = False는 count의 내림차순
```

```python
from matplotlib import pyplot as plt
```

```python
#차트 그리기

pd.Series(x1).value_counts(sort=False).plot(kind = 'bar')

# 왜도 구하기

print(stats.skew(x1))

```

```python
pd.Series(x2).value_counts(sort=False).plot(kind = 'bar')
print(stats.skew(x2))
```

```python
pd.Series(x3).value_counts(sort=False).plot(kind = 'bar')
print(stats.skew(x3))
```

```python
#왜도 계산
print(stats.skew(x1))
print(stats.skew(x2))
print(stats.skew(x3))
```

## 첨도 계산

```python
x1 = [1] * 20 + [2] * 20 + [3] * 20 + [4] * 20 + [5] * 20 # 전혀 뾰족하지 않음
x2 = [1] * 10 + [2] * 20 + [3] * 40 + [4] * 20 + [5] * 10 # 조금 뾰족
x3 = [1] * 2 + [2] * 15 + [3] * 66 + [4] * 15 + [5] * 2 # 매우 뾰족
```

```python
pd.Series(x1).value_counts(sort = False).plot(kind = 'bar')
```

```python
pd.Series(x2).value_counts(sort = False).plot(kind = 'bar')
```

```python
pd.Series(x3).value_counts(sort = False).plot(kind = 'bar')
```

```python
print(stats.kurtosis(x1))
print(stats.kurtosis(x2))
print(stats.kurtosis(x3))
```
